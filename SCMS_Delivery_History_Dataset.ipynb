# Delivery Analysis
## Histogram

````python
df.head(1)
````

###
Extreme delivery delay values were filtered to reduce the impact of outliers and improve visualization clarity.

````python
#delivery delay which customer got late delivery problem
df['delivery_delay'] = (df['delivered_to_client_date'] - df['scheduled_delivery_date']).dt.days

delay_count = df['delivery_delay'].value_counts().reset_index()
delay_count.columns = ['delivery_delay', 'count']
delay_count = delay_count.sort_values(by = 'count', ascending= False)
````

### Removing Outliers from Delivery Delay

Based on the observation of extreme values like -300 and 100 days, we'll filter the `delivery_delay` column to include only values within a more reasonable range, specifically between -100 and 100 days. This will help in creating a more representative visualization of delivery patterns.

````python
# Filter out extreme delivery delay outliers
filtered_delay_df = df[(df['delivery_delay'] >= -20) & (df['delivery_delay'] <= 20)]

# Recalculate delay_count for the filtered data
filtered_delay_count = filtered_delay_df['delivery_delay'].value_counts().reset_index()
filtered_delay_count.columns = ['delivery_delay', 'count']
filtered_delay_count = filtered_delay_count.sort_values(by='count', ascending=False)

# Plot the histogram with filtered data
fig_filtered = px.histogram(filtered_delay_count, x='delivery_delay', y='count', title='Delivery Delay Distribution (Outliers Removed)')
fig_filtered.update_layout(bargap=0.1)
fig_filtered.show()
````

### Insights from Delivery Delay Distribution (Outliers Removed)

1.  **On-time Deliveries are Most Frequent:** The most common delivery delay is 0-9 days, indicating that a significant majority of deliveries are made exactly on schedule.
2.  **Early Deliveries are Common:** There's a notable number of deliveries arriving earlier than scheduled (negative delay days), suggesting efficient logistics in many cases.
3.  **Few Significant Late Deliveries:** While some deliveries are late (positive delay days), their frequency is considerably lower compared to on-time or early deliveries within the filtered range.
4.  **Concentration Around Zero:** The distribution is heavily concentrated around zero, with counts rapidly decreasing as the delay (either positive or negative) increases, reinforcing the overall efficiency of the delivery process.

# Overall Conclusion of Data Analysis

This exploratory data analysis of the SCMS Delivery History Dataset has provided significant insights into the logistics, procurement, and distribution patterns of medical supplies. Key findings include:

1.  **Data Cleaning and Preprocessing**: Initial data cleaning involved handling missing values in 'Shipment Mode', 'Dosage', 'Weight (Kilograms)', and 'Freight Cost (USD)'. Date columns were converted to datetime objects, with invalid entries set to `NaT` to preserve potential business meaning, rather than discarding critical information. Unnecessary columns were dropped, and column names were standardized for clarity.

2.  **Shipment Mode Dominance**: 'Air' transport is the most frequently used shipment mode, followed by 'Truck', highlighting a preference for faster delivery methods across various countries. 'Ocean' and 'Air Charter' constitute smaller portions of the overall shipments.

3.  **Vendor Landscape**: 'SCMS from RDC' is an overwhelmingly dominant supplier, responsible for a massive proportion of the `line_item_quantity`. This indicates a highly centralized procurement strategy or significant dependency on this particular vendor. The top 3 vendors account for a substantial share of the total quantity, pointing to a concentrated vendor base.

4.  **Delivery Efficiency**: The analysis of `delivery_delay` revealed that a large majority of deliveries are either on schedule or arrive early, suggesting generally efficient logistics. While some delays occur, extreme outliers were infrequent, and the distribution is heavily concentrated around zero delay.

5.  **Product Group Dynamics**: 'ARV' (Antiretrovirals) is the most supplied product group by quantity, significantly surpassing all others. However, 'HRDT' (HIV Rapid Diagnostic Tests) products have the highest average pack price, while 'ARV's have the lowest. This inverse relationship between quantity and unit price is crucial for understanding cost structures and budgeting.

6.  **Weight vs. Shipping Cost**: A general positive correlation exists between `weight_kg` and `shipping_cost`, as expected. However, country-specific variations indicate that other factors like geographical location, infrastructure, and regional logistics networks play a significant role in determining final shipping expenses.

7.  **Brand Quantity Insights**: The 'Generic' brand dominates in terms of `line_item_quantity`, reinforcing the high demand and procurement of generic medications. A few other key brands like 'Determine' and 'Aluvia' also contribute substantially, but there's a steep drop-off in quantity for other brands.

8.  **Dosage Form Analysis**: 'Injection', 'Test kit - Ancillary', and 'Test kit' dosage forms exhibit the highest average unit prices, likely due to their specialized nature. Conversely, 'Oral suspension' and 'Powder for oral solution' have the lowest unit prices. In terms of quantity, 'Tablet - FDC' and 'Tablet' forms are overwhelmingly dominant, demonstrating their widespread use and supply volume. This again shows a clear inverse relationship: high-volume dosage forms tend to have lower unit costs.

Overall, the dataset reveals a supply chain heavily focused on ARV products and generic medications, with a largely efficient delivery system despite the complexities of international logistics. The varying cost and quantity profiles across product groups, brands, and dosage forms provide a foundational understanding for optimizing procurement, inventory management, and distribution strategies.
